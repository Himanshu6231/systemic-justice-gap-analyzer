```python
df = pd.read_csv("justice_gap_engineered_dataset.csv")

print("Dataset Shape:", df.shape)
df.head()
```

```python
for col in df.columns:
    df[col] = df[col].astype(str)
```

```python
transactions = df.apply(
    lambda row: [f"{col}={row[col]}" for col in df.columns],
    axis=1
)
```

```python
transactions.iloc[0]
```

```python
['CASE_TYPE=Criminal', 'REGION=Delhi', 'COMPLEXITY=High', 'DELAY_FLAG=Yes', 'OUTCOME=Unfavorable']
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

encoded_df = pd.DataFrame(te_array, columns=te.columns_)
```

```python
encoded_df.head()
```

```python
joblib.dump(te, "transaction_encoder.pkl")
```

```python
print("Encoded shape:", encoded_df.shape)
print("Sample columns:", encoded_df.columns[:10])
```

```python
MIN_SUPPORT = 0.025
```

```python
frequent_itemsets = fpgrowth(
    encoded_df,
    min_support=MIN_SUPPORT,
    use_colnames=True
)
```

```python
print("Number of frequent itemsets:", frequent_itemsets.shape[0])
frequent_itemsets.head(10)
```

```python
frequent_itemsets.sort_values(
    by="support",
    ascending=False,
    inplace=True
)


```

```python
frequent_itemsets.head(10)
```

```python
frequent_itemsets.to_csv("frequent_itemsets.csv", index=False)
```

```python
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.5
)
```

```python
print("Total rules generated:", rules.shape[0])
rules.head()
```

```python
rules_filtered = rules[rules["lift"] >= 1.3]

print("Rules after lift filtering:", rules_filtered.shape[0])
```

```python
def is_justice_gap(consequents):
    for item in consequents:
        if item == "DELAY_FLAG=Yes" or item == "OUTCOME=Unfavorable":
            return True
    return False

justice_rules = rules_filtered[
    rules_filtered["consequents"].apply(is_justice_gap)
]

print("Justice-gap rules:", justice_rules.shape[0])
```

```python
justice_rules = justice_rules.sort_values(
    by="lift",
    ascending=False
)
```

```python
justice_rules[
    ["antecedents", "consequents", "support", "confidence", "lift"]
].head(10)
```

```python
justice_rules.to_csv("justice_gap_rules.csv", index=False)
joblib.dump(justice_rules, "justice_gap_rules.pkl")
```

```python
base_df = df.copy()
```

```python
def parse_items(itemset):
    """
    Converts frozenset({'A=x','B=y'})  {'A':'x','B':'y'}
    """
    d = {}
    for item in itemset:
        k, v = item.split("=")
        d[k] = v
    return d


def filter_df_by_conditions(df, conditions):
    """
    Filters dataframe based on condition dict
    """
    temp = df.copy()
    for col, val in conditions.items():
        temp = temp[temp[col] == val]
    return temp
```

```python
justice_rules = justice_rules.reset_index(drop=True)

baseline_probs = []
adjusted_lifts = []

for _, row in justice_rules.iterrows():

    antecedents = parse_items(row["antecedents"])
    consequents = parse_items(row["consequents"])

    # Full rule subset
    rule_df = filter_df_by_conditions(base_df, antecedents)

    # Consequent column & value
    cons_col = list(consequents.keys())[0]
    cons_val = consequents[cons_col]

    # Rule confidence (already computed)
    rule_conf = row["confidence"]

    # Baseline: same CASE_TYPE if present, else full dataset
    if "CASE_TYPE" in antecedents:
        baseline_df = base_df[base_df["CASE_TYPE"] == antecedents["CASE_TYPE"]]
    else:
        baseline_df = base_df

    if len(baseline_df) == 0:
        baseline_prob = 0
    else:
        baseline_prob = (
            baseline_df[baseline_df[cons_col] == cons_val].shape[0]
            / baseline_df.shape[0]
        )

    adjusted_lift = rule_conf / baseline_prob if baseline_prob > 0 else 0

    baseline_probs.append(round(baseline_prob, 3))
    adjusted_lifts.append(round(adjusted_lift, 3))
```

```python
justice_rules["baseline_probability"] = baseline_probs
justice_rules["adjusted_lift"] = adjusted_lifts
```

```python
justice_rules[
    ["antecedents", "consequents", "confidence", "lift", "baseline_probability", "adjusted_lift"]
].head(10)
```

```python
def gap_severity(adj_lift):
    if adj_lift >= 1.5:
        return "High"
    elif adj_lift >= 1.2:
        return "Moderate"
    else:
        return "Low"

justice_rules["gap_severity"] = justice_rules["adjusted_lift"].apply(gap_severity)
```

```python
def gap_category(antecedents):
    items = parse_items(antecedents)

    if "REGION" in items:
        return "Regional Disparity"
    elif "COMPLEXITY" in items:
        return "Procedural Bottleneck"
    elif "CASE_TYPE" in items:
        return "Case-Type Bias"
    else:
        return "General Pattern"

justice_rules["gap_category"] = justice_rules["antecedents"].apply(gap_category)
```

```python
justice_rules = justice_rules.sort_values(
    by=["gap_severity", "adjusted_lift"],
    ascending=[True, False]
)
```
